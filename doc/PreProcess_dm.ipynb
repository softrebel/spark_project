# Outlier

```python
outlier_removed_df=df.copy()
for key in ['MMSE','CDR','eTIV','nWBV','ASF','SES','EDUC','Age']:
    max=df['MMSE'].mean() + 3*df['MMSE'].std()
    min=df['MMSE'].mean() - 3*df['MMSE'].std()
    print(f'COLUMN: {key}')
    print(f"Highest allowed MMSE : {max}",)
    print(f"Lowest allowed MMSE : {min}",)
    new_df = df[~((df[key] < max) & (df[key] > min))]
df=outlier_removed_df.copy()
print(df.count)
```

# Dealing with missing values

```python
current_col_frac=(df.isna().sum()/df.count()).to_dict()
current_col_frac
```

Data Count: 373

SES  Missing Count: 19 -> 5 %

MMSE Missing Count: 2 -> 0.5 %

Desired Missing Count: 37 -> 10 %

```python
max_missing_fraction=0.1
# current_col_frac={'SES':0.05,'MMSE':0.005}
for key in df.keys():
    if key == 'Group':
        continue
    current_frac=current_col_frac.get(key,0)
    frac=max_missing_fraction-current_frac
    if frac > 0:
        df.loc[df.sample(frac=frac).index,key] = pd.np.nan
    current_col_frac[key]=frac
df.isna().sum()
```
